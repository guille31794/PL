## Lexer

```python
from sly import Lexer,Parser

global ta = None
global tokenlist = None
global ind = 0
global tabla = {}

class ExLex(Lexer):
    
    ignore = ' \t'
    
    tokens = {ID,TRUE,FALSE,NOT,AND,OR,ASSIGN,PRINT}
    
    literals = {'(',')',';',}
    
    ID = r'[a-zA-Z_][a-zA-Z0-9_]*'
    ID['true'] = TRUE
    ID['false'] = FALSE
    ID['not'] = NOT
    ID['and'] = AND
    ID['or'] = OR
    ID['print'] = PRINT
    
    ASSIGN = r':='
    
    @_(r'\n+')
    def newline(self, t):
        self.lineno += t.value.count('\n')

    def error(self, t):
        print("Illegal character '%s'" % t.value[0])
        self.index += 1
```

## Nodo classes

```python
tabla = {}

class Nodo:
    def evaluar():
        pass
    def toStr():
        pass
    
class NodoTRUE(Nodo):
    def evaluar():
        return True
    def toStr():
        return "true"
    
# Fin de NodoTRUE
    
class NodoFALSE(Nodo):
    def evaluar():
        return False
    def toStr():
        return "false"
    
# Fin de NodoFALSE

class NodoID(Nodo):
    def __init__(self,n):
        self.nombre = n
    def evaluar():
        return tabla[self.nombre]
    def toStr():
        return "ID ("+self.nombre+")"
    
# Fin de NodoID

class NodoNOT(Nodo):
    def __init__(self,n):
        self.s = n
    def evaluar():
        return self.s.evaluar()
    def toStr():
        return "NO ("+self.s.toStr()+")"

# Fin de NodoNOT

class NodoAND(Nodo):
    def __init__(self,n,m):
        self.n = n
        self.m = m
    def evaluar():
        return self.n.evaluar() and self.m.evaluar()
    def toStr():
        return "AND ("+self.n.toStr()+" , "+self.m.toStr()+")"

# Fin de NodoAND

class NodoOR(Nodo):
    def __init__(self,n,m):
        self.n = n
        self.m = m
    def evaluar():
        return self.n.evaluar() or self.m.evaluar()
    def toStr():
        return "OR("+self.n.toStr()+" , "+self.m.toStr()+")"
    
# Fin de NodoOR

class NodoPRINT(Nodo):
    def __init__(self,n):
        self.n = n
    def evaluar():
        temp = self.n.evaluar()
        print("Resultado ",int(temp))
        return temp
    def toStr():
        return ("PRINT ",self.n.toStr())
        
# Fin de NodoPRINT

class NodoASIG(Nodo):
    def __init__(self,n,m):
        self.n = n
        self.m = m
    def evaluar():
        temp = self.m.evaluar()
        tabla[self.n] = temp
        return temp
    def toStr():
        return "ASIG("+self.n.toStr(),self.m.toStr()+")"

# Fin de NodoASIG       


```
