```python
from sly import Lexer,Parser

class CalcLexer(Lexer):
    
    tokens = {NUMBER,EQ,NEQ,EQMAY,EQMEN,AND,NOT}
              #OR,NOT}
    ignore = ' \t'
    literals = {'(',')','.',',',';','+','*','-','/'}
    
    EQ = r'=='
    NEQ = r'!='
    EQMAY = r'>='
    EQMEN = r'<='
    AND = r'&'
    #OR = r'|'
    NOT = r'!'
    
    @_(r'[0-9]+')
    def NUMBER(self,t):
        t.value = int(t.value)
        return t
    
    def error(self, t):
        print('Line %d: Bad character %r' % (self.lineno, t.value[0]))
        self.index += 1
    
class CalcParser(Parser):
    tokens = CalcLexer().tokens
    
    def __init__(self):
        self.names = { }
    
    @_('entrada expr ";"')
    def entrada(self, p):
        print("El resultado es:",p.expr)

    @_(' ')
    def entrada(self, p):
        print(" ")

    @_('expr "+" sum')
    def expr(self, p):
        return p.expr + p.sum
    
    @_('expr "-" sum')
    def expr(self, p):
        return p.expr - p.sum

    @_('sum')
    def expr(self, p):
        return p.sum

    @_('sum "*" factor')
    def sum(self, p):
        return p.sum * p.factor
    
    @_('sum "/" factor')
    def sum(self, p):
        return p.sum / p.factor
    
    @_('factor EQ factor')
    def sum(self,p):
        return int(p.factor0 == p.factor1)
    
    @_('factor NEQ factor')
    def sum(self,p):
        return int(p.factor0 != p.factor1)
    
    @_('factor EQMAY factor')
    def sum(self,p):
        return int(p.factor0 >= p.factor1)
    
    @_('factor EQMEN factor')
    def sum(self,p):
        return int(p.factor0 <= p.factor1)
    
    @_('factor AND factor')   
    def sum(self,p):
        return int(p.factor0 & p.factor1)
    
    #@_('factor OR factor')   
    #def sum(self,p):
    #    return int(p.factor0 | p.factor1)
    
    @_('NOT factor')   
    def sum(self,p):
        return ~p.factor
    
    @_('factor')
    def sum(self, p):
        return p.factor

    @_('"-" factor')
    def factor(self, p):
        return -p.factor

    @_('NUMBER')
    def factor(self, p):
        return p.NUMBER

    @_('"(" expr ")"')
    def factor(self, p):
        return p.expr
    
if __name__ == '__main__':
    lexer = CalcLexer()
    parser = CalcParser()
    while True:
        try:
            text = input('operation >')
        except EOFError:
            break
        if text:
            parser.parse(lexer.tokenize(text))
```
