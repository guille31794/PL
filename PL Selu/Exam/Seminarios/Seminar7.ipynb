```python
from sly import Lexer, Parser
import sys
```

```python
temp=0 #Nos ayudara con los registros temporales.

#Clase para pintar todo lo que se desee.
class emite():
    def f(a):
        print(a)

class CalcLexer(Lexer):
    tokens={ID,NUM}
    literals={'=','+','-','*','/',';'}
    ignore=r'[ \t\n]'
    
    ID=r'[a-zA-Z][A-zA-Z0-9]*'
    NUM=r'[0-9]+'
    
    @_('[a-zA-Z][A-zA-Z0-9]*')
    def ID(self, f):
        f.value=str(f.value)
        return f
    @_('[0-9]+')
    def NUM(self, f):
        f.value=str(f.value)
        return f    
    
    def error(self, f):
        print('Bad character %r' %f.value[0])
        sys.exit(0)
        
class CalcParser(Parser):
    tokens=CalcLexer.tokens
    def __init__(self):
        self.names={}
    
    @_('ID "=" suma ";"')
    def entrada(self, f):
        global temp
        temp=temp+1
        emite.f('T'+str(temp)+'='+'T'+str(temp-1))
        temp=0
        pass
    
    @_('ID "=" ID ";"')
    def entrada(self, f):
        emite.f(f.ID0+'='+f.ID1)
        pass
    
    @_('suma "+" prod emptyOper')
    def suma(self, f):
        global temp
        return 'T'+str(temp)
        
    @_('suma "-" prod emptyOper')
    def suma(self, f):
        global temp
        return 'T'+str(temp)
    
    @_('prod')
    def suma(self, f):
        return f.prod

    @_('prod "*" una emptyOper')
    def prod(self, f):
        global temp
        return 'T'+str(temp)
    
    @_('prod "/" una emptyOper')
    def prod(self, f):
        global temp
        return 'T'+str(temp)
    
    @_('una')
    def prod(self, f):
        return f.una

    @_('')
    def emptyOper(self, f):
        global temp
        temp=temp+1
        emite.f('T'+str(temp)+'='+str(f[-3])+str(f[-2])+str(f[-1]))
        pass
        
    @_('ID empty')
    def una(self, f):
        global temp
        return 'T'+str(temp)
    
    @_('NUM empty')
    def una(self, f):
        global temp
        return 'T'+str(temp)
    
    @_('dos')
    def una(self, f):
        return f.dos
    
    @_('"-" ID emptyUna')
    def dos(self, f):
        global tmp
        return 'T'+str(temp)
    
    @_('')
    def emptyUna(self, f):
        global temp
        temp=temp+1
        emite.f('T'+str(temp)+'='+'-'+str(f[-1]))
        pass    

    @_('')
    def empty(self, f):
        global temp
        temp=temp+1
        emite.f('T'+str(temp)+'='+f[-1])
        pass
    
if __name__=='__main__':
    lexer=CalcLexer()
    parser=CalcParser()
    while True:
        try:
            text=input('one expression plix > ')
        except EOFError:
            break
        if text:
            parser.parse(lexer.tokenize(text))
```
