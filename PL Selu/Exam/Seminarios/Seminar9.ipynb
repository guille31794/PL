```python
from sly import Lexer, Parser
import sys
```

```python
#Tabla donde ira el ID mas su tipo.
tablaT={}
tablaValores={}
tabla={}
tipo=None
tipoPuntero=None

class NodoType():
    def f(a):
        global tipo
        tipo=a
        pass

class NodoPuntero():
    def f():
        global tipoPuntero
        if tipoPuntero!=None:
            tipoPuntero='puntero a '+tipoPuntero
        else:
            tipoPuntero='puntero a '
        pass

class NodoTabla():
    def f(a):
        global tipo, tipoPuntero
        if tipoPuntero!=None:
            tablaT[a]=tipoPuntero+tipo
        else:
            tablaT[a]=tipo    
        tabla[a]=tipo
        tablaValores[a]=None
        tipoPuntero=None
        pass

class NodoAsig():
    def ID(a,b):
        if tabla[b]!=None and a in tabla and b in tabla:
            if tabla[a]==tabla[b]:
                tablaValores[a]=tablaValores[b]
            else:
                print('Tipos diferentes.')
        else:
            print('Error, id no en tabla.')
        pass
    def CTEENT(a,b):    
        if a in tabla:
            if tabla[a]=='int':
                tablaValores[a]=b
            else:
                print('Tipos diferentes.')
        else:
            print('Error, id no en tabla.')
        pass  
    def CTEFLOAT(a,b):
        if a in tabla:
            if tabla[a]=='float':
                tablaValores[a]=b
            else:
                print('Tipos diferentes.')
        else:
            print('Error, id no en tabla.')
        pass       

class CalcLexer(Lexer):
    tokens={INT,FLOAT,CHAR,ID,CTEENT,CTEFLOAT}
    literals={';',',','*','=','.'}
    ignore=r'[ \t\n]'
    
    CTEENT=r'[0-9]+'
    CTEFLOAT=r'(\d+\.\d+)'
    INT=r'int'
    CHAR=r'char'
    FLOAT=r'float'
    ID=r'[a-zA-Z][a-zA-Z0-9]*'
    
    @_('[a-zA-Z][a-zA-Z0-9]*')
    def ID(self, f):
        f.value=str(f.value)
        return f
    
    @_('[0-9]+')
    def CTEENT(self, f):
        f.value=int(f.value)
        return f
    
    @_('(\d+\.\d+)')
    def CTEFLOAT(self, f):
        f.value=float(f.value)
        return f
    
    def error(self, f):
        print('Error en %r.'%f.value[0])
        sys.exit(0)
        pass
    
class CalcParser(Parser):
    tokens=CalcLexer.tokens
    def __init__(self):
        self.names={}
        
    @_('defi ";" entrada')
    def entrada(self, f):
        pass
    
    @_('entradaValores')
    def entrada(self, f):
        pass
    
    @_('')
    def entrada(self, f):
        pass
    
    @_('tipo variables')
    def defi(self, f):
        pass
    
    @_('INT emptyType')
    def tipo(self, f):
        pass
    
    @_('CHAR emptyType')
    def tipo(self, f):
        pass
    
    @_('FLOAT emptyType')
    def tipo(self, f):
        pass
    
    @_('')
    def emptyType(self, f):    
        NodoType.f(f[-1])
        pass
    
    @_('var')
    def variables(self, f):
        pass
    
    @_('puntero ID emptyID "," var')
    def var(self, f):
        pass
    
    @_('puntero ID emptyID')
    def var(self, f):
        pass
    
    @_('')
    def emptyID(self, f):
        NodoTabla.f(f[-1])
        pass

    
    @_('"*" puntero')
    def puntero(self, f):
        NodoPuntero.f()
        pass
    
    @_('')
    def puntero(self, f):
        pass
    
    #ENTRADA VALORES:
    @_('ID "=" CTEENT emptyAsig ";" entradaValores')
    def entradaValores(self, f):
        pass

    @_('ID "=" CTEFLOAT emptyAsig ";" entradaValores')
    def entradaValores(self, f):
        pass
    
    @_('ID "=" ID emptyAsig ";" entradaValores')
    def entradaValores(self, f):
        pass
    
    @_('')
    def entradaValores(self, f):
        print(tablaValores)
        pass
    
    @_('')
    def emptyAsig(self, f):
        if f[-1] in tabla:
            NodoAsig.ID(f[-3],f[-1])
        elif type(f[-1])==int:
            NodoAsig.CTEENT(f[-3],f[-1])
        elif type(f[-1])==float:
            NodoAsig.CTEFLOAT(f[-3],f[-1])
        pass
    
if __name__=='__main__':
    lexer=CalcLexer()
    parser=CalcParser()
    while True:
        try:
            text = input('data type list > ')
        except EOFError:
            break
        if text:
            parser.parse(lexer.tokenize(text))
```
