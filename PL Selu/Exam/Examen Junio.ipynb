# Analizador léxico

```python
class CalcLexer(Lexer):
    tokens = {ID, FLECHA, CTEENT, CTEFLOAT, CARACTER, ENTERO, FLOTANTE}
    ignore = ' \t'
    literals = {';', ',', '+', '&', '*', '=', '[', ']', '(', ')'}
    
    # Regular expression rules for tokens
    ID   = r'[a-zA-Z_][a-zA-Z0-9_]*'
    ID['char']  = CARACTER
    ID['int']   = ENTERO
    ID['float'] = FLOTANTE
    
    FLECHA = r'->'
    
    @_(r'\d+\.\d*')
    def CTEFOLAT(self, t):
        t.value = float(t.value)
        return t
    
    @_(r'\d+')
    def CTEENT(self, t):
        t.value = int(t.value)
        return t
    
    @_(r'\n+')
    def newline(self, t):
        self.lineno += t.value.count('\n')

    def error(self, t):
        print("Illegal character '%s'" % t.value[0])
        sys.exit(0)
```
