```python
from sly import Lexer, Parser
import sys
```

```python
tabla={}

class Nodo():
    def escribir(self):
        pass

class NodoEntero(Nodo):
    def escribir(self):
        return 'Entero'

class NodoFlotante(Nodo):
    def escribir(self):
        return 'Flotante'

class NodoCaracter(Nodo):
    def escribir(self):
        return 'Caracter'

class NodoPuntero(Nodo):
    def __init__(self,n):
        self.n=n
    def escribir(self):
        return 'Puntero a ' + self.n.escribir()

class CalcLexer(Lexer):
    tokens={ID,INT, FLOAT, CHAR}
    literals={',',';','*'}
    ignore=r'[ \t\n]'
    ID=r'[a-zA-Z][a-zA-Z0-9]*'
    ID['char']=CHAR
    ID['int']=INT
    ID['float']=FLOAT
    
    def error(self, f):
        print('Bad character %r' % f.value[0])
        sys.exit(0)
        
class CalcParser(Parser):
    tokens=CalcLexer.tokens
    def __init__(self):
        self.names={}
        
    @_('defi entrada')
    def entrada(self, f):
        pass
        
    @_('')
    def entrada(self, f):
        print(tabla)
        pass
        
    @_('tipo lista ";"')
    def defi(self, f):
        return f.tipo
        
    @_('INT')
    def tipo(self, f):
        return NodoEntero()
        
    @_('FLOAT')
    def tipo(self, f):
        return NodoFlotante()
        
    @_('CHAR')
    def tipo(self, f):
        return NodoCaracter()
    
    @_('empty5 elm empty4 resto')
    def lista(self, f):
        return f[-2]
    
    @_('')
    def empty4(self, f):
        return f[-3]
        
    @_('')
    def empty5(self, f):
        return f[-1]
    
    @_('"," empty3 elm empty2 resto')
    def resto(self, f):
        return f[-4]
    
    @_('')
    def empty2(self, f):
        return f[-4]
        
    @_('')
    def empty3(self, f):
        return f[-2]
    
    @_('')
    def resto(self, f):
        return f[-1]
    
    @_('"*" empty1 elm')
    def elm(self, f):
        return f.empty1
    
    @_('')
    def empty1(self, f):
        return NodoPuntero(f[-2])
    
    @_('ID')
    def elm(self, f):
        tabla[f.ID]=f[-2].escribir()
        pass
    
if __name__ == '__main__':
    lexer = CalcLexer()
    parser = CalcParser()
    while True:
        try:
            text = input('one expression plix > ')
        except EOFError:
            break
        if text:
            parser.parse(lexer.tokenize(text))
```
