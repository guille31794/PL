```python
from sly import Lexer, Parser
import sys

tabla={}

class NodoGeneral():
    def General(a):
        printf(a);
        return a

class NodoInic():
    def Inic(a,b):
        tabla[a]=b
        print(tabla) 
    
class NodoSum():
    def Suma(a,b):
        print('%d + %d' %(a,b))
        return a+b;
    
class NodoResta():
    def Resta(a,b):
        print('%d - %d' %(a,b))
        return a-b
    
class NodoMul():
    def Mul(a,b):
        print('%d * %d' %(a,b))
        return a*b

class NodoDiv():
    def Dic(a,b):
        print('%d / %d' %(a,b))
        return a/b
    
class NodoUna():
    def Una(a):
        print('-%d' %a)
        return -a
    
class NodoTabla():
    def Tabla(a):
        return tabla[a]

class CalcLexer(Lexer):
    tokens = {DIGIT, ID}
    DIGIT = r'[0-9]+'
    ID = r'[a-zA-Z][a-zA-Z0-9]*'
    
    literals = {'+', '-', '*', '/', '(', ')', '=', ';'}
    
    @_(r'[0-9]+')
    def DIGIT(self, f):
        f.value=int(f.value)
        return f
    
    def ID(self, f):
        f.value = str(f.value)
        return f
    
    def error(self, t):
        print('Bad character %r' % t.value[0])
        sys.exit(0)
        
class CalcParser(Parser):
    tokens = CalcLexer.tokens
    def __init__(self):
        self.names = { }
        
    @_('')
    def entrada(self, p):
        pass
    
    @_('entrada ID "=" expr ";"')
    def entrada(self,p):
        NodoInic.Inic(p.ID,p.expr)    
             
    @_('comp')
    def expr(self, p):
        return NodoGeneral.General(p.comp)

    @_('comp "+" sum')
    def comp(self, p):
        return NodoSum.Suma(p.comp, p.sum)
    
    @_('comp "-" sum')
    def comp(self, p):
        return NodoResta.Resta(p.comp, p.sum)
    
    @_('sum')
    def comp(self, p):
        return NodoGeneral.General(p.sum)
    
    @_('sum "*" fact')
    def sum(self, p):
        return NodoMul.Mul(p.sum , p.fact)
    
    @_('sum "/" fact')
    def sum(self, p):
        return NodoDiv.Div(p.sum , p.fact)
    
    @_('fact')
    def sum(self, p):
        return NodoGeneral.General(p.fact)
    
    @_('"-" fact')
    def fact(self, p):
        return NodoUni.Uni(p.fact)
    
    @_('ID')
    def fact(self, p):
        return NodoTabla.Tabla(p.ID)
    
    @_('"(" expr ")"')
    def fact(self, p):
        return NodoGeneral.General(p.expr)
    
    @_('DIGIT')
    def fact(self, p):
        return NodoGeneral.General(p.DIGIT)

if __name__ == '__main__':
    lexer = CalcLexer()
    parser = CalcParser()
    while True:
        try:
            text = input('one expression plix > ')
        except EOFError:
            break
        if text:
            parser.parse(lexer.tokenize(text))
```
