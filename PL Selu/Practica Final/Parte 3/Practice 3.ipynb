# Practice 1: Syntax-directed translation

Design a grammar using a bottom - up translator that accepts the following input in C language:
    - Assignment statement
    - Conditional instructions (if - else)
    - Comparison operators (==, <=, >=, !=)
    - Logic operators (AND, OR, NOT)
    - Arithmetic operators (+, - , *, / and unary operator (-))
    - Variables, numerical constants and vectors

```python
from sly import Lexer, Parser
import sys

tabla={}

class NodoMas():
    def Mas(a):
        if a in tabla:
            print(tabla[a])
        else:
            print("Syntax error: ID '%c' isnt in table." %a) 

class NodoID():
    def ID(a,b):
        tabla[a]=b
        print(tabla)

class NodoGen():
    def Gen(a):
        return a
    
class NodoMe():
    def Me(a,b):
        return a<b

class NodoMa():
    def Ma(a,b):
        return a>b
    
class NodoMeI():
    def MeI(a,b):
        return a<=b

class NodoMaI():
    def MaI(a,b):
        return a>=b

class NodoI():
    def I(a,b):
        return a==b
    
class NodoAnd():
    def And(a,b):
        return a&b

class NodoOr():
    def Or(a,b):
        return a|b

class NodoNot():
    def Not(a):
        return ~a
    
class NodoSum():
    def Sum(a,b):
        return a+b
    
class NodoResta():
    def Resta(a,b):
        return a-b
    
class NodoMul():
    def Mul(a,b):
        return a*b
    
class NodoDiv():
    def Div(a,b):
        return a/b
    
class NodoRU():
    def RU(a):
        return -a
    
class NodoTabla():
    def Tabla(a):
        return tabla[a]

class CalcLexer(Lexer):
    tokens = {DIGIT, ID, LESEQ, BIGEQ, EQ, PRINTF, PGENERIC}
    ignore = ' \t'
    DIGIT = r'[0-9]+'
    PRINTF = r'printf'
    PGENERIC = r'%d'
    ID = r'[a-zA-Z][a-zA-Z0-9]*'
    
    LESEQ = r'<='
    BIGEQ = r'>='
    EQ = r'=='
    
    literals = {'+', '-', '*', '/', '(', ')', '<', '>', '=', ';', '|', '&', '~', ',', '"'}
    
    @_(r'[0-9]+')
    def DIGIT(self, f):
        f.value=int(f.value)
        return f
    
    def ID(self, f):
        f.value = str(f.value)
        return f
    
    def error(self, t):
        print('Bad character %r' % t.value[0])
        sys.exit(0)    
    
class CalcParser(Parser):
    tokens = CalcLexer.tokens
    def __init__(self):
        self.names = { }
        
    @_('')
    def entrada(self, p):
        pass
    
    @_('entrada datos')
    def entrada(self,p):
        pass
    
    @_('PRINTF "(" """ PGENERIC mas ID ")" ";"')
    def datos(self, p):
        NodoMas.Mas(p.ID)
            
    @_('""" ","')
    def mas(self,p):
        pass
    
    @_('PGENERIC mas ID ","')
    def mas(self ,p):
        NodoMas.Mas(p.ID)        
        
    @_('ID "=" expr ";"')
    def datos(self, p):    #Aqui tenemos ya el p.ID.
        NodoID.ID(p.ID, p.expr)
         
    @_('expr "<" comp')
    def expr(self, p):
        return NodoMe.Me(p.expr, p.comp)
    
    @_('expr ">" comp')
    def expr(self, p):
        return NodeoMa.Ma(p.expr, p.comp)
    
    @_('expr LESEQ comp')
    def expr(self, p):
        return NodoMeI.MeI(p.expr, p.comp)
    
    @_('expr BIGEQ comp')
    def expr(self, p):
        return NodeoMaI.MaI(p.expr, p.comp)
    
    @_('expr EQ comp')
    def expr(self, p):
        return NodoI.I(p.expr, p.comp)
    
    @_('expr "&" comp')
    def expr(self, p):
        return NodoAnd.And(p.expr, p.comp)
    
    @_('expr "|" comp')
    def expr(self, p):
        return NodoOr.Or(p.expr, p.comp)
    
    @_('"~" expr')
    def expr(self, p):
        return NodoNot.Not(p.expr)
    
    @_('comp')
    def expr(self, p):
        return NodoGen.Gen(p.comp)

    @_('comp "+" sum')
    def comp(self, p):
        return NodoSum.Sum(p.comp, p.sum)
    
    @_('comp "-" sum')
    def comp(self, p):
        return NodoResta.Resta(p.comp, p.sum)
    
    @_('sum')
    def comp(self, p):
        return NodoGen.Gen(p.sum)
    
    @_('sum "*" fact')
    def sum(self, p):
        return NodoMul.Mul(p.sum, p.fact)
    
    @_('sum "/" fact')
    def sum(self, p):
        return NodoDiv.Div(p.sum, p.fact)
    
    @_('fact')
    def sum(self, p):
        return NodoGen.Gen(p.fact)
    
    @_('"-" fact')
    def fact(self, p):
        return NodoRU.RU(p.fact)
    
    @_('ID')
    def fact(self, p):
        return NodoTabla.Tabla(p.ID)
    
    @_('"(" expr ")"')
    def fact(self, p):
        return NodoGen.Gen(p.expr)
    
    @_('DIGIT')
    def fact(self, p):
        return NodoGen.Gen(p.DIGIT)

if __name__ == '__main__':
    lexer = CalcLexer()
    parser = CalcParser()
    while True:
        try:
            text = input('one expression plix > ')
        except EOFError:
            break
        if text:
            parser.parse(lexer.tokenize(text))
        
```
